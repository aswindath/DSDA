```python
subject_columns = ['math score', 'reading score', 'writing score']
print("ANOVA Results based on Gender:")
for subject in subject_columns:
    anova_result_gender = stats.f_oneway(
        dt[dt['gender'] == 'male'][subject],
        dt[dt['gender'] == 'female'][subject]
    )
    
    if anova_result_gender.pvalue < 0.05:
        print(f"There is statistically significant difference in {subject} scores between genders.")
    else:
        print(f"There is no statistically significant difference in {subject} scores between genders.")

print("ANOVA Results based on Test Preparation Course:")
for subject in subject_columns:
    anova_result_course = stats.f_oneway(
        dt[dt['test preparation course'] == 'completed'][subject],
        dt[dt['test preparation course'] == 'none'][subject]
    )
    
    if anova_result_course.pvalue < 0.05:
        print(f"There is statistically significant difference in {subject} scores based on test preparation course.")
    else:
        print(f"There is no statistically significant difference in {subject} scores based on test preparation course.")
```
