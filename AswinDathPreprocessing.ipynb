```python
#Fill in all the missing values present in all the columns in the dataset
mean_age = dt['Age'].mean()
dt['Age'].fillna(mean_age, inplace=True)

mean_fare = dt['Fare'].mean()
dt['Fare'].fillna(mean_fare, inplace=True)

mode_embarked = dt['Embarked'].mode()[0]
dt['Embarked'].fillna(mode_embarked, inplace=True)

dt['Cabin'].fillna('Unknown', inplace=True)
dt['Deck'] = dt['Cabin'].str.extract('([A-Za-z])')
dt['Deck'].fillna('UnknownDeck', inplace=True)
```

```python
dt.isnull().sum()
```

```python
#Check and handle outliers in at least 3 columns in the dataset
plt.figure(figsize=(12, 4))

plt.subplot(131)
dt.boxplot(column='Age')
plt.title('age')

plt.subplot(132)
dt.boxplot(column='Fare')
plt.title('fare')


plt.subplot(133)
dt.boxplot(column='SibSp')
plt.title('sibsp')

plt.show()
```

```python
upper_bound = dt['Fare'].quantile(0.95)
lower_bound = dt['Fare'].quantile(0.05)
dt['Fare'] = dt['Fare'].clip(lower_bound, upper_bound)

upper_bound = dt['Age'].quantile(0.95)
lower_bound = dt['Age'].quantile(0.05)
dt['Age'] = dt['Age'].clip(lower_bound, upper_bound)

upper_bound = dt['SibSp'].quantile(0.95)
lower_bound = dt['SibSp'].quantile(0.05)
dt['SibSp'] = dt['SibSp'].clip(lower_bound, upper_bound)
```

```python
dt.head()
```

```python
#Do min max scaling on the feature set (Take ‘Survived’ as target)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = dt.select_dtypes(include=['number']).drop(['Survived'], axis=1)
y = dt['Survived']
X_scaled = scaler.fit_transform(X)
dt_scaled = pd.DataFrame(X_scaled, columns=X.columns)
dt_scaled['Survived'] = y
```
